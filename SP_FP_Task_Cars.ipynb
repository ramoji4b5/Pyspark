```python
schema = StructType()\
    .add("symboling",IntegerType(),True)\
    .add("normalized-losses",IntegerType(),True)\
    .add("make",StringType(),True)\
    .add("fuel-type",StringType(),True)\
    .add("aspiration",StringType(),True)\
    .add("num-of-doors",StringType(),True)\
    .add("body-style",StringType(),True)\
    .add("drive-wheels",StringType(),True)\
    .add("engine-location",StringType(),True)\
    .add("wheel-base",DoubleType(),True)\
    .add("length",DoubleType(),True)\
    .add("width",DoubleType(),True)\
    .add("height",DoubleType(),True)\
    .add("curb-weight",IntegerType(),True)\
    .add("engine-type",StringType(),True)\
    .add("num-of-cylinders",StringType(),True)\
    .add("engine-size",IntegerType(),True)\
    .add("fuel-system",StringType(),True)\
    .add("bore",DoubleType(),True)\
    .add("stroke",DoubleType(),True)\
    .add("compression-ratio",DoubleType(),True)\
    .add("horsepower",DoubleType(),True)\
    .add("peak-rpm",DoubleType(),True)\
    .add("city-mpg",IntegerType(),True)\
    .add("highway-mpg",IntegerType(),True)\
    .add("price",DoubleType(),True)
```

```python
Automobile_df_with_schema = spark.read.format("csv") \
      .option("header", True) \
      .schema(schema) \
      .load("Automobile_data.csv")
```

# Count of Missing values of dataframe in pyspark using isnan() Function

```python
from pyspark.sql.functions import isnull
#Automobile_df_with_schema.isnull().sum()
null_columns = [c for c in Automobile_df_with_schema.columns if Automobile_df_with_schema.filter(isnull(c)).count() > 0]

print(null_columns)
```

```python
from pyspark.sql.functions import isnan, when, count, col
Automobile_df_with_schema.select([count(when(isnan(c), c)).alias(c) for c in Automobile_df_with_schema.columns]).show()
```

# Count of null values of dataframe in pyspark using isnull() Function

```python
from pyspark.sql.functions import isnan, when, count, col
Automobile_df_with_schema.select([count(when(col(c).isNull(), c)).alias(c) for c in Automobile_df_with_schema.columns]).show()
```

```python
import pyspark.sql.functions as F
import pyspark.sql.types as T
from pyspark.sql.functions import col, avg
final_data_mpg = final_data_df.withColumn("mpg",(F.col("city-mpg")+F.col("highway-mpg"))/2)
final_data_mpg.show()
```

```python
#final_data_mpg.registerTempTable("final_data_mpg")
final_data_mpg.createOrReplaceTempView("automobile_mpg")
```

```python
spark.sql("show tables").show()
```

```python
spark.sql("select make,mpg from automobile_mpg order by mpg desc limit 10").show()
```

```python
spark.sql("select make, price from automobile_mpg order by price desc limit 10").show()
```

```python
def saftey_level(value):
  if   value > 0: 
      return 'Risky'
  elif value == 0:
      return 'Neutral'
  else :
    return 'Safe'
        
#convert to a UDF Function by passing in the function and return type of function
safteyLevelfunc = F.udf(saftey_level, StringType())
final_data_saftey = final_data_mpg.withColumn("Saftey", safteyLevelfunc("symboling"))
final_data_saftey.show()
```

```python
#final_data_saftey.registerTempTable("final_data_saftey")
final_data_saftey.createOrReplaceTempView("automobile_saftey")
```

```python
def car_Class(value):
    if (value >= 25000): 
        return 'TopClass'
    elif (value >= 10000 and  value < 25000):
        return 'Mid-Class'
    else :
        return 'BudgetClass'
        
#convert to a UDF Function by passing in the function and return type of function
carClassfunc = F.udf(car_Class, StringType())
final_data_car_class = final_data_saftey.withColumn("title", carClassfunc("price"))
final_data_car_class.show()
```

```python
final_data_car_class.registerTempTable("automobile_car_class")
```

```python
spark.sql("select distinct title, count(title) as count from automobile_car_class  group by title order by count asc").show()
```

```python
spark.sql("select avg(mpg) as avgMpg, min(mpg) as minMpg, max(mpg) as maxMpg from automobile_car_class").show()
```
